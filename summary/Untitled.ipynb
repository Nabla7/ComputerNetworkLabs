```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes and edges
edges = [
    ("1. Internets network layer", "2. IPv4 & IPv6 addresses"),
    ("2. IPv4 & IPv6 addresses", "3. IP address components"),
    ("3. IP address components", "4. DHCP"),
    ("4. DHCP", "5. NAT"),
    ("6. Router forwarding", "7. Routing algorithms"),
    ("7. Routing algorithms", "8. Centralized vs distribution routing"),
    ("8. Centralized vs distribution routing", "9. Routing levels"),
    ("9. Routing levels", "10. Intra-AS routing"),
    ("10. Intra-AS routing", "11. Inter-AS routing"),
    ("12. Router queueing", "13. Queuing policies"),
]
G.add_edges_from(edges)

# Add standalone nodes
G.add_node("14. SDN")
G.add_node("15. Network management protocols")

# Draw the graph
plt.figure(figsize=(10,10))
pos = nx.spring_layout(G, seed=7)  # positions for all nodes - you might want to seed for reproducibility
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1500, edge_cmap=plt.cm.Blues, font_size=10, arrowsize=20)

plt.show()
```
